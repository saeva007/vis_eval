# PM10 S2 Ordinal Paper Eval

This notebook runs the ordinal cumulative evaluation pipeline for `PMST_net_test_13_s2_pm10_ordinal.py`.

It is a thin wrapper around `paper_eval.run_paper_eval_pm10_ordinal` so that configuration stays here while evaluation logic stays in Python.

```python
from pathlib import Path
import subprocess
import sys

BASE = Path('/public/home/putianshu/vis_mlp')
EXP_ID = 'exp_1773802782_pm10_ordinal_cumulative'
OUTPUT_DIR = 'paper_eval_results_exp_1773802782_pm10_ordinal_cumulative'
DATA_DIR = BASE / 'ml_dataset_s2_tianji_12h_pm10_monthtail_2'
CKPT_DIR = BASE / 'checkpoints'
FOG_TH = 0.50
LOW_VIS_TH = 0.50
BATCH_SIZE = 8192
```

```python
cmd = [
    sys.executable,
    '-m',
    'paper_eval.run_paper_eval_pm10_ordinal',
    '--exp-id', EXP_ID,
    '--output-dir', OUTPUT_DIR,
    '--data-dir', str(DATA_DIR),
    '--ckpt-dir', str(CKPT_DIR),
    '--fog-th', str(FOG_TH),
    '--low-vis-th', str(LOW_VIS_TH),
    '--batch-size', str(BATCH_SIZE),
]
print(' '.join(cmd))
subprocess.run(cmd, cwd=str(BASE), check=True)
```
